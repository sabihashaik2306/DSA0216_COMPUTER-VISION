```python
# =========================================
# Region-Based Segmentation with Output Display
# =========================================

!pip install opencv-python-headless -q  # Install OpenCV in Colab

import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Step 1: Upload your image
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Get the uploaded file name

# Step 2: Read the image in grayscale
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Step 3: Apply Gaussian blur to remove noise
blur = cv2.GaussianBlur(img, (5, 5), 0)

# Step 4: Apply simple thresholding for region segmentation
# Pixels > 127 -> 255 (white), Pixels <=127 -> 0 (black)
_, segmented = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)

# Step 5: Display original and segmented images
print("Original Image:")
cv2_imshow(img)
print("Region-Segmented Output:")
cv2_imshow(segmented)

# Step 6: Save the segmented image
output_file = "region_segmented.jpg"
cv2.imwrite(output_file, segmented)

# Step 7: Download the segmented image
files.download(output_file)
```
