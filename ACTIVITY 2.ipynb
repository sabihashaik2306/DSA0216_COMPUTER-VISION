```python
# =========================================
# Contour Detection in Google Colab
# =========================================

# Install OpenCV (for Colab)
!pip install opencv-python-headless -q

import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# Step 1: Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Automatically get uploaded file name

# Step 2: Read the uploaded image
image = cv2.imread(filename)

# Step 3: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 4: Apply Gaussian Blur
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Step 5: Edge detection
edges = cv2.Canny(blur, 50, 150)

# Step 6: Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Step 7: Draw contours
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

# Step 8: Show the final contour image
cv2_imshow(image)

# Step 9: Save and download output
output_filename = "output_contours.jpg"
cv2.imwrite(output_filename, image)
files.download(output_filename)
```
